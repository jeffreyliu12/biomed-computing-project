```python
import pandas as pd
df = pd.read_csv('COVID-19_Case_Surveillance_Public_Use_Data_with_Geography_20260203.csv')
```

```python
df
```

```python
df = df.dropna(subset=['res_county', 'county_fips_code'])
df
```

```python
df_covid_infections_by_county = df.groupby(['res_county', 'county_fips_code']).size().reset_index(name='covid_infections')

df_covid_infections_by_county = df_covid_infections_by_county.sort_values(by='covid_infections', ascending=False)
df_covid_infections_by_county
```
