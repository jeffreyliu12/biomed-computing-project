```python
import pandas as pd
import numpy as np
import matplotlib as plt
import zipfile

zf = zipfile.ZipFile('COVID-19_Case_Surveillance_Public_Use_Data_with_Geography_20260203.csv.zip')
df = pd.read_csv(zf.open('COVID-19_Case_Surveillance_Public_Use_Data_with_Geography_20260203.csv'), low_memory=False)
```

```python
df
```

```python
df = df.dropna(subset=['res_county', 'county_fips_code'])
df_2022 = df[df['case_month'].str.startswith('2022')].copy()
```

```python
df_covid_infections_by_county_2022 = df_2022.groupby(['res_county', 'county_fips_code']).size().reset_index(name='covid_infections')

df_covid_infections_by_county_2022 = df_covid_infections_by_county_2022.sort_values(by='covid_infections', ascending=False)
df_covid_infections_by_county_2022
```
